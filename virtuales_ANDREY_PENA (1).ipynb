```python
import pandas as pd

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00352/Online%20Retail.xlsx'
data = pd.read_excel(url)

print(data.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

print('Valores faltantes por columna:\n', data.isnull().sum())
print('\nEstad√≠sticas descriptivas:\n', data.describe())

data = data.dropna(subset=['CustomerID'])
data = data[data['Quantity'] > 0]

plt.figure(figsize=(10, 8))
sns.countplot(y='Country', data=data, order=data['Country'].value_counts().index[:15])
plt.title('Top 15 Pa√≠ses por N√∫mero de Transacciones')
plt.xlabel('N√∫mero de Transacciones')
plt.ylabel('Pa√≠s')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import altair as alt

# Evitar vegafusion: forzar transformador por defecto y permitir grandes tablas
alt.data_transformers.disable_max_rows()
alt.data_transformers.enable('default')
alt.renderers.enable('default')

# Preparaciones y descomposici√≥n de series
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])
daily_sales = data.set_index('InvoiceDate')['Quantity'].resample('D').sum()
decomposition = sm.tsa.seasonal_decompose(daily_sales.dropna(), model='additive')
fig = decomposition.plot()
plt.tight_layout()
plt.show()

# Filtrado y tipos
data = data.dropna(subset=['Quantity', 'UnitPrice'])
data = data[(data['Quantity'] > 0) & (data['UnitPrice'] > 0)]
data['InvoiceNo'] = data['InvoiceNo'].astype(str)

# Usar una muestra para Altair (ajustable)
sample = data.sample(min(2000, len(data)), random_state=42)

# Gr√°fico de dispersi√≥n
scatter_chart = (
    alt.Chart(sample)
    .mark_circle(size=60, color='steelblue', opacity=0.6)
    .encode(
        x=alt.X('Quantity:Q', title='Cantidad'),
        y=alt.Y('UnitPrice:Q', title='Precio Unitario'),
        tooltip=[alt.Tooltip('InvoiceNo:N'), alt.Tooltip('StockCode:N'),
                 alt.Tooltip('Quantity:Q'), alt.Tooltip('UnitPrice:Q'),
                 alt.Tooltip('Country:N')]
    )
    .properties(title='Gr√°fico de dispersi√≥n: Cantidad vs. Precio Unitario (Online Retail)',
                width=600, height=400)
    .interactive()
)

scatter_chart
```

```python
import pandas as pd
from dash import Dash, dcc, html, Input, Output
import plotly.express as px

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00352/Online%20Retail.xlsx'
data = pd.read_excel(url)
data = data.dropna(subset=['Quantity','UnitPrice','Country','Description'])
data = data[(data['Quantity']>0) & (data['UnitPrice']>0)]
paises = sorted(data['Country'].unique())

app = Dash(__name__)
app.layout = html.Div([
    html.H1('Top 10 productos m√°s vendidos por pa√≠s'),
    dcc.Dropdown(id='dropdown-pais', options=[{'label': pais, 'value': pais} for pais in paises], value='United Kingdom', clearable=False, style={'width':'50%'}),
    dcc.Graph(id='grafico-productos')
])

@app.callback(Output('grafico-productos','figure'), Input('dropdown-pais','value'))
def actualizar_grafico(pais):
    df = data[data['Country']==pais]
    top = df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10).reset_index()
    fig = px.bar(top, x='Description', y='Quantity', color='Quantity', color_continuous_scale='Blues')
    fig.update_layout(xaxis_tickangle=-45)
    return fig

# if __name__ == '__main__':
#     app.run(debug=True)
```
